# Wordcloud

Word clouds (also known as text clouds or tag clouds) work in a simple way: the more a specific word appears in a source of textual data (such as a speech, blog post, or database), the bigger and bolder it appears in the word cloud.

Python package already exists in Python for generating `word` clouds. The package, called `word_cloud` was developed by **Andreas Mueller**. You can learn more about the package by following this [link](https://github.com/amueller/word_cloud/).

Let's use this package to learn how to generate a word cloud for a given text document.


```python
# # import package and its set of stopwords
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

##### Open alice.txt to create wordcloud in this demo

```python
with open('alice.txt', 'r') as f:
    alice = f.read()
```

```python
alice[0:500]
```

##### Generate Simple Wordclould

```python
# instantiate a word cloud object
wc = WordCloud(
     background_color='white',
     width=1000, height=600, # resolution
     stopwords=STOPWORDS,
     max_words=2000)

# generate the word cloud
wc.generate(alice)
```

```python
# display the word cloud
fig = plt.figure(figsize=(10, 6))
plt.imshow(wc)
plt.axis('off')
plt.show()
```

##### Stopwords

The default `STOPWORDS` is a python set that includes common words which provide little to no value to the meaning of the text.

**said** isn't really an informative word. So let's add it to our stopwords and re-generate the cloud.

```python
# add 'said' to STOPWORDS
STOPWORDS.add('said')
print(STOPWORDS)
```

 ##### Mask
 
You can superimpose the words onto a mask of any shape with the word_cloud package. Let's use a mask alice_mask.png.

```python
# save mask to alice_mask
alice_mask = np.array(Image.open('alice_mask.png'))
```

##### Generate Wordclould with Mask

```python
# instantiate a word cloud object
wc = WordCloud(
     background_color='white',
     width=1400, height=1800, # resolution
     stopwords=STOPWORDS,
     max_words=2000, 
     mask=alice_mask)

# generate the word cloud
wc.generate(alice)

# display the word cloud
fig = plt.figure(figsize=(14, 18))
plt.imshow(wc)
plt.axis('off')
plt.show()
```
