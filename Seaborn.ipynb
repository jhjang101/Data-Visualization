# Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative **statistical graphics**.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Axes-level and Figure-level functions

The plotting functions in seaborn are broadly divided into two classes

* "Axes-level" functions plots data onto a single matplotlib.pyplot.Axes object, including `regplot`, `boxplot`, `kdeplot`, and many others

* "Figure-level" functions interface with matplotlib through a seaborn object, including `relplot`, `catplot`, `displot`, `pairplot`, `jointplot` and one or two others

##### Read data

```python
df_im = pd.read_csv('immigration.csv')
df_im.head()
```

```python
df_ex = pd.read_csv('exams.csv')
df_ex.head()
```

##### Set seaborn style theme

Seaborn has five built-in themes to style its plots: **darkgrid**(default) , **whitegrid** , **dark** , **white** , and **ticks** 

```python
sns.set_theme()
```

##### Axes-level plot - single column

set x and y column for single column plot. 

```python
# plot immigrants from asia vs year

sns.lineplot(data=df_im, x='Year',y='Asia')
plt.title('Immigration Trend')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
sns.regplot(data=df_ex, x='math score', y='reading score');
```

##### Axes-level plot - multi columns

Set index to x axis using `.set_index()`

```python
sns.lineplot(data=df_im.set_index('Year'));
```

```python
# No need to set index as the gender column contains string values
sns.boxplot(data=df_ex)
```

##### Figure-level plot - single column

set x and y column for single column plot. 

```python
f = sns.relplot(data=df_im, x='Year',y='Asia', kind='line')

# Title for the complete figure
f.fig.suptitle("Immigrant from Asia" )

# Set x-axis and y-axis labels
f.set_axis_labels( 'Year' , 'Total Immigrant' )
```

```python
sns.jointplot(data=df_ex, x='math score', y='reading score', kind='scatter')
```

```python
sns.displot(data=df_ex, x='math score', kind='hist')
```

##### Figure-level plot - multi column

Set index to x axis using `.set_index()`

```python
f = sns.relplot(data=df_im.set_index('Year'), kind='line',
            height=4, aspect=2)


# Title for the complete figure
f.fig.suptitle('Immigrant')

# Set x-axis and y-axis labels
f.set_axis_labels( 'Year' , 'Total Immigrant' )
```

```python
sns.catplot(data=df_ex, col='gender', kind='violin' )
```

```python
sns.jointplot(data=df_ex.set_index('writing score'), kind='scatter')
```

```python
sns.displot(data=df_ex.set_index('writing score'), kind='kde')
```

```python
sns.pairplot(data=df_ex, hue='gender')
```
